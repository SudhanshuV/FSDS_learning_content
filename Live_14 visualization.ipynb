```python
x=np.linspace(-1,6,30)
```

```python
x
```

```python
y=np.linspace(-1,6,30)
```

```python
y
```

```python
def sin_fun(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
z=sin_fun(x,y)
```

```python
z
```

```python
ax=plt.axes(projection='3d')
ax.plot3D(x,y,z)
```

```python
z=x+y
```

```python
ay=plt.axes(projection='3d')
ay.plot3D(x,y,z)
ay.set_xlabel('x')
ay.set_ylabel('y')
ay.set_zlabel('z')
```
